#**Lane Chosen: Content Refresh / Decay Prediction**

**Why**: It is highly actionable. Instead of guessing which articles or search pages are losing traffic, we can use a machine learning model to proactively identify high-value, declining pages so content creators can update them before traffic completely bottoms out.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**The Search/Business Question**: "Which of our highly-visible pages are currently on a downward traffic trend and require immediate content refreshing?"

**The Unit of Analysis**: A single page (URL).

**The Output**: A probability score (between 0 and 1) representing the likelihood that the page's search traffic is declining.

**The Decision/Action**: A content editor reviews the top 20 flagged pages weekly and schedules them for a content update (rewriting, updating statistics, or adding new info).

**The Cost of a Wrong Call**:

**False Positive (Flagging a healthy page)**: Editors waste valuable hours rewriting content that didn't need updating.

**False Negative (Missing a declining page)**: A key traffic-generating page continues to decay, resulting in a permanent loss of search engine visibility and ad revenue.

**Why ML helps (instead of just a manual check)**: Keeping track of hundreds or thousands of pages manually is impossible. Simple rules like "check everything older than 6 months" flag too many pages. Machine learning looks at multi-dimensional patterns (impressions, age, click-through rates, and position) to prioritize the pages truly in need of rescue.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

total_pages = df.shape[0]
declining_rate = df["is_declining_label"].mean()
high_visibility_stale = df[(df["impressions_90d"] >= 500) & (df["days_since_last_update"] >= 180)].shape[0]

print(f"Total pages in starter dataset: {total_pages}")
print(f"Overall page declining rate: {declining_rate:.1%}")
print(f"Number of stale, high-visibility pages: {high_visibility_stale}")
```
